```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
city_ride_combined = pd.merge(city_data, ride_data, on="city")
# Display the data table for preview
city_ride_combined.head()
```

## Bubble Plot of Ride Sharing Data

```python
test = city_data.loc[(city_data["city"] == 'Barajasview')]
test1 = city_ride_combined.loc[(city_ride_combined["city"] == 'Barajasview')]
test1
```

```python
# Obtain the x and y coordinates for each of the three city types
urban = city_ride_combined.loc[city_ride_combined["type"] == "Urban"].groupby("city")
suburban = city_ride_combined.loc[city_ride_combined["type"] == "Suburban"].groupby("city")
rural = city_ride_combined.loc[city_ride_combined["type"] == "Rural"].groupby("city")

# # Build the scatter plots for each city types
urban_fare = urban["fare"].mean()
urban_rides = urban["ride_id"].count()
urban_driver_count = urban["driver_count"].sum()
plt.scatter(marker = 'o', x=urban_rides, y=urban_fare, alpha = 0.75, color = "#D95319", edgecolor = "black", s = urban_driver_count)

suburban_fare = suburban["fare"].mean()
suburban_rides = suburban["ride_id"].count()
suburban_driver_count = suburban["driver_count"].sum()
plt.scatter(marker = 'o', x=suburban_rides, y=suburban_fare, alpha = 0.75, color = "lightblue", edgecolor = "black", s=suburban_driver_count)

rural_fare = rural["fare"].mean()
rural_rides = rural["ride_id"].count()
rural_driver_count = rural["driver_count"].sum()
plt.scatter(marker = 'o', x=rural_rides, y=rural_fare, alpha = 0.75, color = "yellow", edgecolor = "black", s = rural_driver_count)

# # Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number Of Rides Per City")
plt.ylabel("Average Fare ($)")
# Create a legend
labels = ["Urban", "Suburban", "Rural"]
lgnd = plt.legend(labels)

# reference : https://stackoverflow.com/questions/24706125/setting-a-fixed-size-for-points-in-legend
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
plt.grid()
# Incorporate a text label regarding circle size
# Reference : https://stackoverflow.com/questions/42435446/how-to-put-text-outside-python-plots
plt.text(0.925, 0.5, "Note:\n", fontsize=10, transform=plt.gcf().transFigure)
plt.text(0.925, 0.490, "Circle size corelates with driver count per city", fontsize=10, transform=plt.gcf().transFigure)
# Save Figure
plt.savefig("../Images/scatterplot.png")
plt.show()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_fare = city_ride_combined["fare"].sum()
urban_fare = city_ride_combined.loc[city_ride_combined["type"] == "Urban"]["fare"].sum()
suburban_fare = city_ride_combined.loc[city_ride_combined["type"] == "Suburban"]["fare"].sum()
rural_fare = city_ride_combined.loc[city_ride_combined["type"] == "Rural"]["fare"].sum()
urban_percent = (urban_fare/total_fare) * 100
suburban_percent = (suburban_fare/total_fare) * 100
rural_percent = (rural_fare/total_fare) * 100

# # Build Pie Chart
# # Labels for the sections of our pie chart
labels = ["Urban", "Rural", "Suburban"]

# The values of each section of the pie chart
sizes = [urban_percent, rural_percent, suburban_percent]

# The colors of each section of the pie chart
colors = ["lightcoral", "yellow", "lightskyblue"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0.1, 0, 0)

plt.pie(sizes, labels=labels, colors=colors, explode = explode, shadow=True, autopct="%1.1f%%", startangle=270)
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("../Images/FarePie.png")
plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_rides = city_ride_combined["ride_id"].count()
urban_rides = city_ride_combined.loc[city_ride_combined["type"] == "Urban"]["ride_id"].count()
suburban_rides = city_ride_combined.loc[city_ride_combined["type"] == "Suburban"]["ride_id"].count()
rural_rides = city_ride_combined.loc[city_ride_combined["type"] == "Rural"]["ride_id"].count()
urban_ride_percent = (urban_rides/total_rides) * 100
suburban_ride_percent = (suburban_rides/total_rides) * 100
rural_ride_percent = (rural_rides/total_rides) * 100

# # Build Pie Chart
# # Labels for the sections of our pie chart
labels = ["Urban", "Rural", "Suburban"]

# The values of each section of the pie chart
sizes = [urban_ride_percent, rural_ride_percent, suburban_ride_percent]

# The colors of each section of the pie chart
colors = ["lightcoral", "yellow", "lightskyblue"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0.1, 0, 0)

plt.pie(sizes, labels=labels, colors=colors, explode = explode, shadow=True, autopct="%1.1f%%", startangle=260)
plt.title("% of Total Rides by City Type")
# # Save Figure
plt.savefig("../Images/RidePie.png")
plt.show()

```

## Total Drivers by City Type

```python
# Calculate Driver Percents
total_drivers = city_data["driver_count"].sum()
urban_drivers = city_data.loc[city_data["type"] == "Urban"]["driver_count"].sum()
suburban_drivers = city_data.loc[city_data["type"] == "Suburban"]["driver_count"].sum()
rural_drivers = city_data.loc[city_data["type"] == "Rural"]["driver_count"].sum()

urban_driver_percent = (urban_drivers/total_drivers) * 100
suburban_driver_percent = (suburban_drivers/total_drivers) * 100
rural_driver_percent = (rural_drivers/total_drivers) * 100

urban_driver_percent, suburban_driver_percent, rural_driver_percent
# Build Pie Charts
labels = ["Urban", "Rural", "Suburban"]

# The values of each section of the pie chart
sizes = [urban_driver_percent, rural_driver_percent, suburban_driver_percent]

# The colors of each section of the pie chart
colors = ["lightcoral", "yellow", "lightskyblue"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0.1, 0, 0)

plt.pie(sizes, labels=labels, colors=colors, explode = explode, shadow=True, autopct="%1.1f%%", startangle=220)
plt.title("% of Total Drivers by City Type")
# Save Figure
plt.savefig("../Images/DriverPie.png")
plt.show()
```
